**¿Cuáles han sido los departamentos (TOP 3) más afectados a nivel histórico y en el último año?**

Comparativas de cada reporte 

```python
#Secuestros por departamento a nivel historico
g_dep_secuestrost = df_secuestros.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_secuestrost = g_dep_secuestrost.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Secuestros Historicamente"})

g_dep_secuestrost.head(3)
```

```python
ua_secuestros = df_secuestros.loc[df_secuestros["FECHA HECHO"] >= '01/09/2020']
g_dep_secuestros = ua_secuestros.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_secuestros = g_dep_secuestros.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Secuestros Ultimo año"})
g_dep_secuestros.head(3)
```

```python
pd.concat([g_dep_secuestrost,g_dep_secuestros], axis =1).head(3)
```

```python
#Para secuestros los 3 departamentos mas afectados fueron,VALLE, CUNDINAMARCA, y ANTIOQUIA tanto historicamente como en el ultimo año
```

```python
#Abusos por departamento a nivel historico
g_dep_abusost = df_abusos.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_abusost = g_dep_abusost.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Abusos Historicamente"})

g_dep_abusost.head(3)
```

```python
ua_abusos = df_abusos.loc[df_abusos["FECHA HECHO"] >= '01/09/2020']
g_dep_abusos = ua_abusos.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_abusos = g_dep_abusos.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Abusos Ultimo año"})
g_dep_abusos.head(3)

```

```python
pd.concat([g_dep_abusost,g_dep_abusos], axis =1).head(3)
```

```python
#Para Abusos los 3 departamentos mas afectados fueron, CUNDINAMARCA, ANTIOQUIA y VALLE tanto historicamente como en el ultimo año
```

```python
#Homicidios por departamento a nivel historico
g_dep_homicidiost = df_homicidios.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_homicidiost = g_dep_homicidiost.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Homicidios Historicamente"})
g_dep_homicidiost.head(3)

```

```python
ua_homicidios = df_homicidios.loc[df_homicidios["FECHA HECHO"] >= '01/09/2020']
g_dep_homicidios = ua_homicidios.groupby("DEPARTAMENTO")["CANTIDAD"].sum().to_frame()
g_dep_homicidios = g_dep_homicidios.sort_values(by = "CANTIDAD", ascending = False).rename(columns = {"CANTIDAD" : "Homicidios Ultimo año"})
g_dep_homicidios.head(3)
```

```python

pd.concat([g_dep_homicidiost,g_dep_homicidios], axis =1).head(3)
```

```python
#Para homicidios los 3 departamentos mas afectados fueron, VALLE, ANTIOQUIA y CUNDINAMARCA tanto historicamente como en el ultimo año
```

TOTAL DAÑO CAUSADO POR DEPARTAMENTO

```python
#aqui unimos las tablas que representan el dano total causado historicamente en cada departamento 
secyabus = pd.merge(g_dep_secuestrost,g_dep_abusost, how = "outer", left_index = True, right_index = True)
tabla_histo = pd.merge(secyabus,g_dep_homicidiost, how = "outer", left_index = True, right_index = True)
tabla_histo["TOTAL DAÑO CAUSADO"] = tabla_histo.sum(axis=1)
tabla_histo.sort_values(by = "TOTAL DAÑO CAUSADO", ascending = False).head(3)
```

```python
#aqui unimos las tablas que representan el dano causado en el ultimo ano en cada departamento 
secyabuso = pd.merge(g_dep_secuestros,g_dep_abusos, how = "outer", left_index = True, right_index = True)
tabla_ua = pd.merge(secyabuso,g_dep_homicidios, how = "outer", left_index = True, right_index = True)
tabla_ua["TOTAL DAÑO CAUSADO"] = tabla_ua.sum(axis=1)
tabla_ua.sort_values(by = "TOTAL DAÑO CAUSADO", ascending = False).head(3)
```

```python
#CONCLUSION
#El top 3 de departamentos mas afectados a nivel historico, en los ultimos 10 anos, son CUNDINAMARCA, seguido por VALLE, y ANTIOQUIA de tercero.
#A nivel del ultimo ano, siendo los mismos departamentos, pero esta vez, ANTIOQUIA de segundo, seguido por VALLE. tanto de manera general en
#Colombia como a nivel especifico, puesto que los 3 departamentos fueron los mas afectados en cada uno de los reportes
```






**¿Cuál ha sido históricamente el arma o medio más común para cometer el delito?**



```python
armas_secuestro = df_secuestros["ARMAS MEDIOS"].value_counts().to_frame().rename(columns = {"ARMAS MEDIOS":"Armas para secuestro"})
armas_abuso = df_abusos["ARMAS MEDIOS"].value_counts().to_frame().rename(columns = {"ARMAS MEDIOS":"Armas para abuso sexual"})
armas_homicidio = df_homicidios["ARMAS MEDIOS"].value_counts().to_frame().rename(columns = {"ARMAS MEDIOS":"Armas para homicidio"})
```

```python
armas_secuestro.head(3).plot(kind="barh") #El arma mas usada para realizar el secuestro es Arma Blanca o Cortopunzante
```

```python
armas_abuso.head(3).plot(kind="barh")#En el abuso sexual lo mas comun es no emplear armas
```

```python
armas_homicidio.head(3).plot(kind="barh") #El arma mas usada para realizar los homicidios es el arma de fuego
```

```python
tabla_armas = pd.merge(armas_secuestro,armas_abuso, how = "outer", left_index = True, right_index = True)
tabla_armas = pd.merge(tabla_armas,armas_homicidio, how = "outer", left_index = True, right_index = True)
tabla_armas["TOTAL USADAS"] = tabla_armas.sum(axis=1)
tabla_armas.sort_values(by = "TOTAL USADAS", ascending = False).head(3).plot(kind="barh")
```

```python
#CONCLUSION
#Historicamente, el arma o medio mas comun para cometer el delito es el arma de fuego.
#Seguido del arma de fuego, es muy popular para los delincuentes no emplern armas para cometer el delito, o que las mismas no sean reportadas
```

**¿Cómo es la proporción de géneros y grupos etarios que han estado involucrados en este tipo de delito? ¿Han variado con el paso de los años?**

```python
df_secuestros['YEAR'] = df_secuestros['FECHA HECHO'].dt.year
generosxtiempo_secuestro = df_secuestros.groupby(["YEAR","GENERO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_secuestro.plot(kind="bar")
generosxtiempo_secuestro 

#Para empezar por los generos, aunque los valores obviamente varian, la proporcion de secuestros siempre ha sido practicamente
#la misma, pues los hombres parecen verse mas afectados por este tipo de delitos desde hace 10 anos, aunque parece hacer disminuido
#desde el 2015
```

```python
df_abusos['YEAR'] = df_abusos['FECHA HECHO'].dt.year
generosxtiempo_abusos = df_abusos.groupby(["YEAR","GENERO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_abusos.plot(kind="bar")
generosxtiempo_abusos
#Por otro lado, aunque los valores obviamente varian, la proporcion de abuso sexual siempre ha sido practicamente
#la misma, pues las mujeres parecen verse mas afectados por este tipo de delitos desde hace 10 anos, ademas de que parecio
#aumentar de manera significativa desde el 2015
```

```python
df_homicidios['YEAR'] = df_homicidios['FECHA HECHO'].dt.year
generosxtiempo_homicidios = df_homicidios.groupby(["YEAR","GENERO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_homicidios.plot(kind="bar")
generosxtiempo_homicidios

# la proporcion de homicidios siempre ha sido practicamente la misma, pues los hombres parecen verse mas afectados
#por este tipo de delitos desde hace 10 anos, y estan en bastante diferencia con los reportes hacia la mujer o aquellos que no reportan
```

```python
generosxtiempo_secuestro = df_secuestros.groupby(["YEAR","GRUPO ETARIO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_secuestro.plot(kind="bar")
generosxtiempo_secuestro 

#Para comparar ahora con respecto a los grupos etarios, tambien parece tener la misma proporcion, los adultos siempre
#viendose mas afectados con los secuestros, seguidos por los menores.
```

```python
generosxtiempo_abusos = df_abusos.groupby(["YEAR","GRUPO ETARIO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_abusos.plot(kind="bar")
generosxtiempo_abusos
#De la misma manera, vemos que en el tema de abusos sexuales, los menores son quienes se ven mas afectados, seguidos por
#los adultos, y la proporcion se ha mantenido igual, desde el 2015 se ve un aumento significativo en todos los grupos etarios
#pero principalmente en menores
```

```python
generosxtiempo_homicidios = df_homicidios.groupby(["YEAR","GRUPO ETARÍO"])["CANTIDAD"].sum().to_frame()
generosxtiempo_homicidios.plot(kind="bar")
generosxtiempo_homicidios

#Para terminar, podemos ver como la proporcion entre homicidios tampoco varia mucho, parece ser que los adultos son
#quienes se ven mas afectados de una manera significativa
```

**¿Se evidencia alguna tendencia para cometer dicho delito en algún mes particular del año?**


```python
df_secuestros['MONTH'] = df_secuestros['FECHA HECHO'].dt.month
secuestro_mes = df_secuestros.groupby(["MONTH"])["CANTIDAD"].sum().to_frame().rename(columns={"CANTIDAD":"SECUESTROS"})
secuestro_mes = secuestro_mes.sort_values(by=["SECUESTROS"], ascending = False)
secuestro_mes.plot(kind="bar")

#Marzo es el mes donde mas secuestros se ven, seguido de agosto y mayo
```

```python

df_abusos['MONTH'] = df_abusos['FECHA HECHO'].dt.month
abuso_mes = df_abusos.groupby(["MONTH"])["CANTIDAD"].sum().to_frame().rename(columns={"CANTIDAD":"ABUSOS"})
abuso_mes = abuso_mes.sort_values(by=["ABUSOS"], ascending = False)
abuso_mes.plot(kind="bar")

#Mayo es el mes donde mas abusos sexuales se ven con una significativa diferencia, seguido de marzo y enero. 

```

```python
df_homicidios['MONTH'] = df_homicidios['FECHA HECHO'].dt.month
homicidio_mes = df_homicidios.groupby("MONTH")["CANTIDAD"].sum().to_frame().rename(columns={"CANTIDAD":"HOMICIDIOS"})
homicidio_mes = homicidio_mes.sort_values(by=["HOMICIDIOS"], ascending = False)
homicidio_mes.plot(kind="bar")
#Los meses con mas homicidios son diciembre, seguido de enero y de tercero mayo
```
